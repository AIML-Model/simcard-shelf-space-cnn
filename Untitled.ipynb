```python
from ultralytics import YOLO

from IPython.display import display, Image
image_path = "/Users/ganesh/Desktop/Git-kyndryl/CAPSTONE-PROJECT/All-images/image-74.jpg"
```

```python
import cv2
image = cv2.imread(image_path)
```

```python
model = YOLO("/Users/ganesh/Desktop/Git-kyndryl/CAPSTONE-PROJECT/capstone-sim-shelf-space/simcard-shelf-space-2/runs/detect/train/weights/best.pt")
results = model.predict(image, conf=0.25, iou=0.45, imgsz=640)
#class_names = model.names
```

```python
for result in results:
    # Loop through each detected box
    for box in result.boxes:
        # Get the bounding box coordinates (x1, y1, x2, y2)
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        
        # Get class ID and class name
        class_id = int(box.cls[0])  # Class index (ID)
        class_name = model.names[class_id] 
        #class_name = class_names[class_id]  # Class name

        # Calculate bounding box width and height
        box_width = x2 - x1
        box_height = y2 - y1
        
        # Print the class name, bounding box width, and height
        print(f"Class: {class_name} (ID: {class_id})")
        print(f"Bounding Box Width: {box_width}, Bounding Box Height: {box_height}")
        print("------")
```
